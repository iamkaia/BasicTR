```python
import requests 
import bs4
import time
import json
import re
# 把 到 ptt 八卦版 網址存到URL 變數中
base_url = 'https://www.ptt.cc'
URL = "https://www.ptt.cc/bbs/Psychiatry/index.html"
# 設定Header與Cookie
my_headers = {'cookie': 'over18=1;'}
# 發送get 請求 到 ptt 八卦版
#response = requests.get(URL, headers = my_headers)
# 印出回傳網頁程式碼
#print(response.text)

#if response.status_code == 200:
# 2-1 把網頁程式碼(HTML) 丟入 bs4模組分析
#soup = bs4.BeautifulSoup(response.text,"html.parser")
# 2-2 查找所有html 元素 過濾出 標籤名稱為 'div' 同時class為 title 
#titles = soup.find_all('div',class_='title')

# 2-3 萃取文字出來。
# 因為我們有多個Tags存放在 List titles中。
# 所以需要使用for 迴圈將逐筆將List
#file_path = 
articles_data = []
pages_to_crawl = 5

# 解析單篇文章內容
def get_article_content(article_url):
    response = requests.get(article_url)
    if response.status_code == 200:
        soup = bs4.BeautifulSoup(response.text,"html.parser")
        # 2-2 查找所有html 元素 過濾出 標籤名稱為 'div' 同時class為 title 
        content = soup.find('div',id='main-content')
        #print(content)
        pure_content = re.split(r'2024|※|--',content.text)[1].strip()
        #for content in contents:
        #if contents[1]==True:
        #    print(f'文章內容: {contents[1]}')
        pushs = soup.find_all('span',class_='f3 push-content')
        response_list = []
        for push in pushs:
            push = push.text
            #print(f'推{push}')
            response_list.append(push)
        article_data = {
                "title": article_title,
                "content": pure_content,
                "reponses":response_list
            }
        articles_data.append(article_data)
    return None

# 進行多頁爬取
current_url = URL

for _ in range(pages_to_crawl):
    response = requests.get(current_url, headers = my_headers)
    if response.status_code == 200:
        soup = bs4.BeautifulSoup(response.text,"html.parser")
        #print(soup)
        titles = soup.find_all('div',class_='title')
        for title in titles:
            #print(title)
            link_tags = title.find_all('a')
            for link_tag in link_tags:
                # 提取文章標題和網址
                article_title = link_tag.text
                article_url = 'https://www.ptt.cc' + link_tag['href']
                print(f'文章標題: {article_title}, 文章鏈接: {article_url}')
                article_content = get_article_content(article_url)
        # 獲取上一頁的鏈接（跳轉到下一頁）
        paging = soup.find('a', string='‹ 上頁')
        if paging:
            current_url = base_url + paging['href']
            print(current_url)
        else:
            print("no!!!")
            break  # 如果沒有上一頁，則結束
        
        time.sleep(1)
'''
def serialize_article(article):
    # 如果 article 中有 'Tag' 或其他自定义对象，将它转换为字符串或其他可序列化的格式
    if isinstance(article, tag):
        return str(article)  # 或者 article.text，取决于你想保存的内容
    raise TypeError(f"Object of type {article.__class__.__name__} is not JSON serializable")
'''

# 將結果寫入 JSON 檔
with open('ptt_articles.json', 'w', encoding='utf-8') as f:
    json.dump(articles_data, f, ensure_ascii=False, indent=4, default=serialize_article)
```
